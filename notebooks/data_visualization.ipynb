### Global Visualization of Seismic Events

```python
import polars as pl
import plotly.express as px
import plotly.io as pio
from src.constants import SESMIC_DATA_CSV

pio.renderers.default = "notebook_connected"

df = pl.read_csv(SESMIC_DATA_CSV)
df_clean = df.drop_nulls(subset=["latitude", "longitude", "mag"])


plot_data = df_clean.sample(n=50000) if len(df_clean) > 50000 else df_clean

fig = px.density_mapbox(
    plot_data.to_pandas(),
    lat='latitude',
    lon='longitude',
    z='mag',
    radius=8,
    center=dict(lat=0, lon=0),
    zoom=1,
    mapbox_style="open-street-map",
    title="Global Seismic Event Density (Jupyter Render)",
    opacity=0.3
)

fig.update_layout(margin={"r":0,"t":40,"l":0,"b":0})
fig.show()
```
